```python
# librerias
import numpy as np
import pandas as pd
```

```python
# lectura del dataset Employeee.ipynb de https://www.kaggle.com/datasets/tejashvi14/employee-future-prediction
empleados = pd.read_csv('Employee.csv')
print(empleados)

# Elimina registros duplicados
empleados = empleados.drop_duplicates()
print(empleados)
```

```python
# Calcula si corresponde la media, mediana y moda. De cada campo del dataset.

# NIVEL DE EDUCACIÓN
moda_educacion = empleados['Education'].mode().values[0]

# AÑO DE INGRESO A LA EMPRESA
media_ingreso = round(empleados['JoiningYear'].mean())
mediana_ingreso = round(empleados['JoiningYear'].median())

# OFICINA: CIUDAD DONDE SE ENCUENTRA
moda_oficina = empleados['City'].mode().values[0]

# NIVEL DE PAGO
media_pago = int(empleados['PaymentTier'].mean())
mediana_pago = int(empleados['PaymentTier'].median())

# EDAD ACTUAL
media_edad = round(empleados['Age'].mean())
mediana_edad = round(empleados['Age'].median())

# GÉNERO DEL EMPLEADO
moda_genero = empleados['Gender'].mode().values[0]

# ALGUNA VEZ SE HA QUEDADO FUERA DE PROYECTOS DURANTE 1 MES O MÁS
moda_queda_fuera = empleados['EverBenched'].mode().values[0]

# EXPERIENCIA EN CAMPO ACTUAL
media_experiencia = round(empleados['ExperienceInCurrentDomain'].mean(),2)
mediana_experiencia = int(empleados['ExperienceInCurrentDomain'].median())

# SI EL EMPLEADO DEJA LA EMPRESA EN LOS PRÓXIMOS 2 AÑOS
media_abandono = round(empleados['LeaveOrNot'].mean(),2)
mediana_abandono = int(empleados['LeaveOrNot'].median())

# Imprimir los resultados
print("Moda de nivel de educación:", moda_educacion)
print()
print("Media de año de ingreso:", media_ingreso)
print("Mediana de año de ingreso:", mediana_ingreso)
print()
print("Moda de la ubicacion de la oficina:", moda_oficina)
print()
print("Media del nivel de pago:", media_pago)
print("Mediana del nivel de pago:", mediana_pago)
print()
print("Media de la edad del empleado/a:", media_edad)
print("Mediana de la edad del empleado/a:", mediana_edad)
print()
print("Moda del genero del empleado/a:", moda_genero)
print()
print("Moda si quedo afuera de un proyecto por más de un mes:", moda_queda_fuera)
print()
print("Media de experiencia en el campo:", media_experiencia)
print("Mediana de experiencia en el campo:", mediana_experiencia)
print()
print("Media de renuncia en los próximos 2 años:", media_abandono)
print("Mediana de erenuncia en los próximos 2 años:", mediana_abandono)
print()
```
